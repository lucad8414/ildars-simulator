```python
import numpy as np
from objects import Vector, Point, Line

def angle(self, other: Vector) -> float:
        return np.arccos(self.direction.dot(other) / (abs(self.direction) * abs(other)))
```

This is the Formula of angle:
$$
\alpha = \cos^{-1} \left(\frac{\vec{a} \cdot \vec{b}}{||\vec{a}|| \cdot ||\vec{b}||} \right) \quad \text{iff} \quad \vec{a} \neq \begin{pmatrix} 0 \\ 0 \end{pmatrix} \quad \text{and} \quad \vec{b} \neq \begin{pmatrix} 0 \\ 0 \end{pmatrix}
                                                                                                                    
$$
 

```python
from typing import Optional
def intersection(self, other: Point | Vector | "Line") -> Optional[Point]:
        if isinstance(other, Vector):
            ext = other.extend()
        
        elif isinstance(other, Line):
            ext = other
        
        y = (self.direction[1] * (self.anchor[0] - ext.anchor[0]) + self.direction[0] * (ext.anchor[1] - self.anchor[1])) / (self.direction[0] * ext.direction[1] - self.direction[1] * ext.direction[0])
        # Reversing the vector is not allowed, so we decline that.
        if y > 0:
            return ext.point(y)
        else:
            return None
```

Getting the intersection Point of two Vectors with following structure:
$$
\vec{v} = \begin{pmatrix} p_0 \\ p_1 \end{pmatrix} + x \cdot \begin{pmatrix} r_0 \\ r_1 \end{pmatrix}; \qquad 
\vec{u} = \begin{pmatrix} q_0 \\ q_1 \end{pmatrix} + y \cdot \begin{pmatrix} s_0 \\ s_1 \end{pmatrix}
$$
Now we solve after $x$ and $y$.
$$
\begin{align*}
\vec{v} &= \vec{u} \\
\begin{pmatrix} p_0 \\ p_1 \end{pmatrix} + x \cdot \begin{pmatrix} r_0 \\ r_1 \end{pmatrix} &= \begin{pmatrix} q_0 \\ q_1 \end{pmatrix} + y \cdot \begin{pmatrix} s_0 \\ s_1 \end{pmatrix} \\
p_0 + x \cdot r_0 &= q_0 + y \cdot s_0 \\
p_1 + x \cdot r_1 &= q_1 + y \cdot s_1 
\end{align*}
\newline
$$
First we solve one equation after $x$, depending on $y$:
$$
\begin{align*}
x = \frac{q_1 - p_1}{r_1} + \frac{s_1y}{r_1} \\
\end{align*}
\newline
$$
We plug the equation into the other one and solve after $y$: 
$$
\begin{align*}
p_0 + \left(\frac{q_1 - p_1}{r_1} + \frac{s_1y}{r_1}\right) \cdot r_0 &= q_0 + y \cdot s_0 \\
p_0 - q_0 + \frac{r_0(q_1 - p_1)}{r_1} + \frac{r_0s_1y}{r_1} &= y \cdot s_0 \\
p_0 - q_0 + \frac{r_0(q_1 - p_1)}{r_1} &= y \cdot s_0 - \frac{r_0s_1y}{r_1} \\
r_1(p_0 - q_0) + r_0(q_1 - p_1) &= y(r_1s_0 - r_0s_1) \\
y = \frac{r_1(p_0 - q_0) + r_0(q_1 - p_1)}{r_1s_0 - r_0s_1} &= \frac{r_1(p_0 - q_0) + r_0(q_1 - p_1)}{det\begin{pmatrix} s_0 & s_1 \\ r_0 & r_1 \end{pmatrix}}
\end{align*}
$$
We only want $y$, because self in this case is the wall in my current implementation.
and we want the vector to be extended, so other should be amplified, until it hits the wall.
this is why we only take values for $y > 0$ into account.
